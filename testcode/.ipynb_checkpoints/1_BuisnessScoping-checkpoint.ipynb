## Distribuition of Funded vs. not funded invdividuals

SQL Query below:

```python

with ds.connect('dm',u,p) as con:
    sql_query = """
    

            SELECT LEFT(Fullapp_First_Submitted_Date_Key,6) as Period, Approved_apps, funded_count_Msr, count(*) as Freq
            FROM [PBdm].[resl].[vw_FactSalesFunnel_Daily]
            where Full_Apps = 1 and left(Fullapp_First_Submitted_Date_Key,6) > 201704 
                AND left(Fullapp_First_Submitted_Date_Key,6) < 201710
            GROUP BY LEFT(Fullapp_First_Submitted_Date_Key,6), Approved_apps, funded_count_Msr

        """
    df1 = con.read_sql(sql_query)

df1.to_csv('tables/2_fundedVSnotFunded.csv', index =False)
```

````python
df1 = pd.read_csv('tables/2_fundedVSnotFunded.csv')
````

````python
df2 = df1[df1.Approved_apps ==1]
df2
````

````python
df3 = pd.pivot_table(df2, index = 'Period', columns='funded_count_Msr', values ='Freq')
df3.plot.bar(title = "Volume of individuals that were funded vs not funded per period")
df3
````

# How can we identify customers that fund with competition vs those that do nothing?

1. Find custnums of indivdiuals that fund with TD --> run through credit beaurea and find logic to identify them
1. Then obtain custnums of individuals that do not fund with TD --> run through CB table and find out what happens to them

```python

with ds.connect('dm',u,p) as con:
    q3 = """
    
        SELECT RIGHT(cust_num, 9) as cust_num, Funded_date_key, *
        FROM [Pbdm].[resl].[vw_Applications_DAily]
        WHERE cust_num is not null AND left(Funded_date_key,6) > 201704 AND left(Funded_date_key,6) < 201710
            AND FullApp_Ind =1 
            AND Approved_Ind = 1 AND Funded_Ind =1 AND Channel =50

        """
    df_fundedCust = con.read_sql(q3)
        
df_fundedCust.to_csv('tables/3_fundedCust.csv', index =False)
```

````python
df_fundedCust = pd.read_csv('tables/3_fundedCust.csv')
````

````python
# Here is a list of all custnums of individuals that got approved

list = df_fundedCust['cust_num'].tolist()
fundedandapprovedcustnums = str(tuple(list))
````

### Pass this list to CB table and obtain their credit limits & No. of MTG's
```python
with ds.connect('dm',u,p) as con:
    q4 = """
    
        SELECT Date_Key, cust_num, count(Prod_Ind) AS NoOfMTGs, sum(limit) as TotalLimit
        FROM [PBdm].[dbo].[Credit_Bureau]
        where cust_num IN %s AND prod_ind = 'MTG'
        group by cust_num, Date_Key, Prod_Ind
        
        """

    df4 = con.read_sql(q4 %fundedandapprovedcustnums)

```

````python
#df4.to_csv('tables/4_df4.csv', index =False)
df4 = pd.read_csv('tables/4_df4.csv')
````

````python
df5 = pd.pivot_table(df4, values = ['NoOfMTGs', 'TotalLimit'], index =['cust_num'], columns =['Date_Key'])
````
